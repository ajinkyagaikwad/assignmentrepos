```
###1. Exception handling
```

```
while True:
	try:
		a = int(input("Enter a number to be divided by 5: "))
		result = 5/a
	except:
		print("ERROR. Please enter non-zero number")
	else:
		print("Computation is: ",result)
		break
```

```
###2. Print all permutations
```

```
x=["Americans","Indians"]
y=["Play", "watch"]
z=["Baseball","cricket"]
for i in range(len(x)):
	for j in range(len(y)):
		for k in range(len(z)):
			print(x[i],y[j],z[k])
```
