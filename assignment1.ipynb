```
###1.	Print numbers divisible by 7, but not 5
```

```
a = 2000
b  =3200
for i in range(a,b+1):
	if (i%7==0) and (i%5!=0):
		print(i,end=',')
```

```
###2.	Reversed sequence of names
```

```
first = input('Enter first name: ')
last = input('Enter last name: ')
print('Reversed result is',end=' ')
print(first[::-1]+' '+last[::-1])
```

```
###3.	Volume of a sphere
```

```
import math
r = 12
volume = (4/3)*(math.pi)*(r**3)
print("%.2f" % volume)
```

```
###4.	Generate list from comma separated integers
```

```
seq = input('Enter comma separated numbers ')
result = seq.split(',')
final = [int(x) for x in result]
print("Desired list is: ", final)
```

```
###5.	Reverse a string
```

```
name = input('Enter a string ')
print('Reversed string is ', name[::-1])
```

```
###6.	Star pattern
```

```
for i in range(1,5):
	for j in range(i):
		print('*',end='')
	print()
for i in range(5,0,-1):
	for j in range(i):
		print('*',end='')
	print()
```
