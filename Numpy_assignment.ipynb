```
import numpy as np
import numpy.polynomial.legendre as leg
```

Moving averages

```
input_list = [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150]
N=3
cumulative_sum, moving_average = [0], []
for i ,x in enumerate(input_list,1):
  cumulative_sum.append(cumulative_sum[i-1]+x)
  if i>=N:
    moving_aver = (cumulative_sum[i] - cumulative_sum[i-N])/N
    moving_average.append("{:.2f}".format(moving_aver))
```

```
print(moving_average)
```
