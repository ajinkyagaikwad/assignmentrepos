```
x = [3,4,5,12,16]
#Custom-built reduce function that works for basic operations: add,subtract,multiply,divide. Similar logic can be used to build solution for other operations too
def myreduce(choice,lst):
	add=0
	sub=0
	mult=1
	div=1
	if choice=='+':
		for i in range(len(lst)):
			add += lst[i]
		return add
	if choice=='-':
		for i in range(len(lst)):
			sub -= lst[i]
		return sub
	if choice=='*':
		for i in range(len(lst)):
			mult *= lst[i]
		return mult
	if choice=='/':
		for i in range(len(lst)):
			div /= lst[i]
		return div

a = myreduce('+',x)   #returns reduced sum
b = myreduce('-',x)   #returns reduce subtraction
c = myreduce('*',x)   #returns reduce product
d = myreduce('/',x)   #returns division
print(a,b,c)
print("{:.4f}".format(d))   #formatting the division upto 4 decimal places.
```

```
#Custom-built filter function. It demonstrates operations like filtering out numbers less than 'n'.
l = [3,4,5,6,7,8]
final =[]
def filter_greater(lst):
	n=int(input("Enter minimum threshold "))
	for i in range(len(lst)):
		if lst[i]>n:
			final.append(lst[i])
	return final
def my_filter(choice,lst):
	if choice == '>':
		result=filter_greater(lst)
	return result

out = my_filter('>',l)
print("Filtered list is :",out)
```

```
###Area of a triangle (Class concept)
```

```
import math
class Triangle:
	def __init__(self,side1,side2,side3):
		self.side1=side1
		self.side2=side2
		self.side3=side3
	def area(self):
		s=(self.side1+self.side2+self.side3)/2
		s1=(s-self.side1)*(s-self.side1)*(s-self.side1)
		s1 = s*s1
		return math.pow(s1,0.5)
	
print("Enter 3 sides of a triangle: ")
a = int(input("1->"))
b = int(input("2->"))
c = int(input("3->"))	
A = Triangle(a,b,c)
final = A.area()
print("Area of triangle is {:.4f}".format(final))
```

```
###Vowel detection
```

```
def detect_vowel(char):
	if char in ('a','e','i','o','u','A','E','I','O','U'):
		return True
	else:
		return False

input_char = input('Enter a character: ')
result=detect_vowel(input_char)
if result==True:
  print('A vowel')
else:
  print('Not a vowel')
```

```
###Longest word
```

```
def longestWord(l):
	largest = l[0]
	for i in range(len(l)):
		if i==len(l)-1:
			return largest
			break
		if len(largest) < len(l[i+1]):
			largest=l[i+1]

l = []
while(True):
	char = input("Enter a word. Press 'q' to quit. ")
	if char=='q':
		break
	l.append(char)

final = longestWord(l)
print("Longest word of the list is: ",final)
```
