```
import pandas as pd
import numpy as np
```

```
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```
df
```

```
###1. Change 'FlightNumber' column
```

```
df.loc[1, 'FlightNumber'] = 10055
df.loc[3, 'FlightNumber'] = 10075
```

```
df= df.astype({'FlightNumber': int})
```

```
df
```

```
###2. Creating two columns in place of one 'From_To'
```

```
df['From'] = df['From_To'].apply(lambda x: x.split('_')[0])
df['To'] = df['From_To'].apply(lambda x: x.split('_')[1])
```

```
df
```

```
###3. Capitalising text in newly created columns. Deleting 'From_To' column
```

```
df['From'] = df['From'].apply(lambda x: x.capitalize())
df['To'] = df['To'].apply(lambda x: x.capitalize())
```

```
df.drop('From_To', axis=1, inplace=True)
```

```
cols = df.columns.to_list()
```

```
cols = cols[-2:] + cols[:-2]
```

```
cols
```

```
df = df[cols]
```

```
df
```

```
###4. Processing 'RecentDelays' columns
```

```
df['delay_1'] = df['RecentDelays'].apply(lambda x: np.nan if len(x)==0 else x[0])
df['delay_2'] = df['RecentDelays'].apply(lambda x: np.nan if len(x)<2 else x[1])
df['delay_3'] = df['RecentDelays'].apply(lambda x: np.nan if len(x)<3 else x[2])
```

```
df
```

```
###5. Deleting 'RecentDelays' columns
df.drop('RecentDelays', axis=1, inplace=True)
```
